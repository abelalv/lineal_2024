# Clase de Matrices
### Introducción
Una **matriz** es una disposición rectangular de números en filas y columnas. Se utiliza en diversas disciplinas, incluyendo matemáticas, física y ciencias de la computación.

La notación típica para una matriz es:

$$ A = \begin{pmatrix}
a_{11} & a_{12} & \dots & a_{1n} \\
a_{21} & a_{22} & \dots & a_{2n} \\
\vdots & \vdots & \ddots & \vdots \\
a_{m1} & a_{m2} & \dots & a_{mn} 
\end{pmatrix} $$


### Operaciones Elementales

#### Suma y Resta de Matrices
La suma y resta de matrices se realiza elemento por elemento, siempre que las matrices tengan las mismas dimensiones.

#### Producto de Matrices
El producto de dos matrices se define como la suma de los productos de las filas de la primera matriz con las columnas de la segunda.

$$ (AB)_{ij} = \sum_{k} a_{ik}b_{kj} $$

#### Transposición de Matrices
La transposición de una matriz implica intercambiar sus filas por columnas.

#### Determinante y Matriz Inversa
El determinante es un valor escalar que se puede calcular a partir de una matriz cuadrada. La matriz inversa es aquella que, cuando se multiplica por la matriz original, da como resultado la matriz identidad.

```python
# Importar bibliotecas necesarias
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Definir matrices de ejemplo
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Suma de matrices
suma = A + B
print("Suma de A y B:\n", suma)

# Resta de matrices
resta = A - B
print("Resta de A y B:\n", resta)

# Producto de matrices
producto = np.dot(A, B)
print("Producto de A y B:\n", producto)

# Transposición de una matriz
transpuesta = np.transpose(A)
print("Transpuesta de A:\n", transpuesta)

# Determinante de una matriz
determinante = np.linalg.det(A)
print("Determinante de A:\n", determinante)

# Matriz inversa
inversa = np.linalg.inv(A)
print("Matriz inversa de A:\n", inversa)
```

### Manipulación de Imágenes
Podemos leer una imagen y convertirla en una matriz. Esto es útil en procesamiento de imágenes y visión por computadora.


```python
# Leer una imagen y convertirla en una matriz
imagen = Image.open('figures/figuralineal.jpg').convert('L') # Cambia la ruta a tu imagen
matriz_imagen = np.array(imagen)

# Mostrar la imagen en blanco y negro
plt.imshow(imagen, cmap='gray')
plt.title("Imagen en Blanco y Negro")
plt.show()

# Mostrar la matriz correspondiente
print("Matriz de la imagen:\n", matriz_imagen)
```
